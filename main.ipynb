```python
import pandas as pd
import io
import requests
```

```python
df = pd.DataFrame()

dataset = [
    'https://raw.githubusercontent.com/GiveMeAllYourCats/VAERSVACCINEJUPITER/master/dataset/2020VAERSData.csv',
    'https://raw.githubusercontent.com/GiveMeAllYourCats/VAERSVACCINEJUPITER/master/dataset/2021VAERSData.csv',
]

print('Importing %s datasets' % len(dataset))

for file in dataset:
    df = df.append(pd.read_csv(file,encoding='cp1252'))
#     print('Imported %s' % file)    
        
print('There are %s datasets which have %s combined results' % len(dataset), len(df))
```
